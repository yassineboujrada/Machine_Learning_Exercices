#### ***Q1***-utilisez la fonction ***make_blobs(n_samples,centers, cluster_std,random_state)*** du module ***sklearn.datasets*** pour generer trois classes puis visualiser ces classes 



```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=200,cluster_std=3,centers=3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
colo=[]
for i in y:
    if i==2:
        colo.append("red")
    elif i==1:
        colo.append("green")
    else:
        colo.append("blue")
```

```python
plt.scatter(x[:,0],x[:,1],c=colo)
```

#### ***Q2***  Ecrire un script pour prÃ©dire la classe de x avec l'algorithme du k-means puis visualiser ces trois classe avec les cluster_centers

```python
from sklearn.cluster import KMeans
```

```python
x=np.array(x)
```

```python
model=KMeans(n_clusters=3).fit(x)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("accuracy is :",accuracy_score(y,model.predict(x))*100,"%")
print("mtrix confusion :\n",confusion_matrix(y,model.predict(x)))
```

#### Q3. A quelle classe apartient les trois points suivants: x1=(-2,1); x2(0,1) ;x3=(2.5,5)  

```python
pt_for_predict=[[-2,1],[0,1],[2.5,5]]
y_pred=model.predict(pt_for_predict)
print("pour les points:\nx1({},{}) =>{}\nx1({},{}) =>{}\nx1({},{}) =>{}".format(pt_for_predict[0][0],pt_for_predict[0][1],y_pred[0],
                                                                               pt_for_predict[1][0],pt_for_predict[1][1],y_pred[1],pt_for_predict[2][0],pt_for_predict[2][1],y_pred[2]))
plt.figure(figsize=(14,5))
plt.scatter(x[:,0],x[:,1],c=colo)
plt.scatter([x[0] for x in pt_for_predict],[x[1] for x in pt_for_predict],c="black")
```

```python
from sklearn.datasets import load_iris
```

```python
data_flower=load_iris()
```

```python
X=data_flower.data
Y=data_flower.target
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
model2=KMeans(n_clusters=3).fit(X)
```

```python
print("accuracy is :",accuracy_score(Y,model2.predict(X))*100,"%")
print("mtrix confusion :\n",confusion_matrix(Y,model2.predict(X)))
```

```python
l=[]
for i in range(1,20):
    mdl=KMeans(n_clusters=i)
    mdl.fit(X)
    l.append(mdl.inertia_)
plt.plot(range(1,20),l,c="red")
```
